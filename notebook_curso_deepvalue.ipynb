## 1.3 Conditionals (if / elif / else)

Conditionals let your code make **decisions**. In trading, this is fundamental:
- Is the price below the moving average? → Buy
- Does the loss exceed 12%? → Sell (stop loss)

The structure is:
```python
if condition:
    # runs if the condition is True
elif other_condition:
    # runs if the first was False and this one is True
else:
    # runs if no condition was True
```

**Important:** Python uses **indentation** (4 spaces) to define code blocks, not braces `{}`.

````python
# Example: stop loss and take profit logic
entry_price = 100.0
current_price = 85.0
stop_loss_pct = -0.12       # -12%
take_profit_pct = 0.15      # +15%

# Compute the current return
return_pct = (current_price - entry_price) / entry_price
print(f"Current return: {return_pct:.2%}")

# Make a decision
if return_pct <= stop_loss_pct:
    action = "SELL (Stop Loss)"
elif return_pct >= take_profit_pct:
    action = "SELL (Take Profit)"
else:
    action = "HOLD"

print(f"Decision: {action}")

# Try changing current_price to 120.0 and run again
````

## 1.5 Functions

A **function** is a reusable block of code with a name. Instead of copying and pasting
the same calculation 100 times, you define it once and call it whenever you need it.

**Why does it matter?** In DeepValueAI, every logical piece is a function:
- `download_ohlcv()` — downloads prices
- `compute_technical_features()` — computes indicators
- `predict_proba()` — generates predictions

**Best practices:**
- Each function should do **one thing only**
- The name should describe **what it does** (verb + noun)
- Use `return` to return results (not just `print`)

````python
# Define reusable functions
def calculate_return(entry_price, current_price):
    """Compute the percentage return of an investment."""
    return (current_price - entry_price) / entry_price

def classify_signal(probability, threshold=0.50):
    """Classify a probability as BUY or HOLD based on the threshold."""
    if probability >= threshold:
        return "BUY"
    return "HOLD"

def evaluate_position(entry_price, current_price, stop_loss=-0.12, take_profit=0.15):
    """Evaluate whether a position should be closed."""
    ret = calculate_return(entry_price, current_price)  # reuse the previous function
    if ret <= stop_loss:
        return "SELL (Stop Loss)", ret
    if ret >= take_profit:
        return "SELL (Take Profit)", ret
    return "HOLD", ret

# Use the functions
print(f"AAPL return: {calculate_return(150, 178):.2%}")
print(f"Signal (prob=0.75): {classify_signal(0.75)}")
print(f"Signal (prob=0.30): {classify_signal(0.30)}")

action, ret = evaluate_position(100, 85)
print(f"\nPosition: {action} (return: {ret:.2%})")
````

## 3.1 What is Machine Learning?

**Machine Learning (ML)** is teaching a computer to make decisions
based on historical data, without programming the rules explicitly.

In DeepValueAI, ML works like this:

```
INPUT (34 features)               OUTPUT
────────────────────              ──────
RSI = 25                  ┌───┐
Williams %R = -85         │   │
PE Ratio = 12.5           │ M │──→  Probability: 0.78 (78%)
VIX = 18.2         ──→    │ L │──→  Decision: BUY
Fed Rate = 5.25           │   │
Sentiment = 0.65          └───┘
...34 features total
```

**Classification vs Regression:**
- **Classification** (what we use): Will it rise or not? → Answer: Yes (1) or No (0)
- **Regression**: How much will it rise? → Answer: +12.5%

We use classification because it's more robust — predicting direction is easier than predicting the exact number.

## 3.2 Prepare the Data

**Golden rule of ML:** NEVER train with data from the future.

In finance this is critical. If you train with data from 2024 and then evaluate with data from 2023,
the model has already "seen the future" — this is called **look-ahead bias** and produces false results.

**The solution:** Temporal split.
```
|────── Train (70%) ──────|── Val (15%) ──|── Test (15%) ──|
2020                      2023            2024             2025
```

````python
from core.config import FEATURE_COLUMNS

# Prepare the data exactly as DeepValueAI does
df_sorted = df.sort_values("date").reset_index(drop=True)

# Replace infinities with NaN (some indicators can produce inf)
df_sorted[FEATURE_COLUMNS] = df_sorted[FEATURE_COLUMNS].replace([np.inf, -np.inf], np.nan)

# Temporal split: 70% train, 15% validation, 15% test
n = len(df_sorted)
train_end = int(n * 0.70)
val_end = int(n * 0.85)

train = df_sorted.iloc[:train_end]
val = df_sorted.iloc[train_end:val_end]
test = df_sorted.iloc[val_end:]

train_range = f"{train['date'].min().date()} → {train['date'].max().date()}"
val_range = f"{val['date'].min().date()} → {val['date'].max().date()}"
test_range = f"{test['date'].min().date()} → {test['date'].max().date()}"
print(f"Train: {len(train):>10,} rows ({train_range})")
print(f"Val:   {len(val):>10,} rows ({val_range})")
print(f"Test:  {len(test):>10,} rows ({test_range})")

# Separate features (X) from the target variable (y)
X_train = train[FEATURE_COLUMNS].values
y_train = train["label"].values
X_val = val[FEATURE_COLUMNS].values
y_val = val["label"].values
X_test = test[FEATURE_COLUMNS].values
y_test = test["label"].values

print(f"\nFeatures: {len(FEATURE_COLUMNS)}")
print(f"Shape of X_train: {X_train.shape}  (rows × features)")
print(f"Balance in train: {y_train.mean():.2%} positive")
````

## 3.3 Train a Model

We'll train the same model DeepValueAI uses: **HistGradientBoostingClassifier**.

**Why this model?**
- Handles NaN natively (many features have missing data)
- Fast to train (millions of rows in minutes)
- It's an sklearn implementation equivalent to LightGBM/XGBoost
- Requires no preprocessing (normalization, imputation)

**How does it work?** It builds many small "decision trees", where each
new tree corrects the errors of the previous one. It's like a team where each member
learns from the others' mistakes.

````python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import (
    f1_score,
    precision_score,
    recall_score,
    roc_auc_score,
)

# Train the model (that's it! 3 lines)
model = HistGradientBoostingClassifier(
    max_iter=200,            # number of trees
    max_depth=5,             # maximum depth of each tree
    learning_rate=0.05,      # how fast it learns (lower = more careful)
    min_samples_leaf=100,    # minimum samples per leaf (prevents overfitting)
    random_state=42,         # seed for reproducibility
)

print("Training model...")
model.fit(X_train, y_train)
print("Model trained!")

# Predict probabilities on the validation set
y_val_proba = model.predict_proba(X_val)[:, 1]  # probability of class 1 (rises)

print("\nFirst 10 predicted probabilities:")
for i in range(10):
    real = "RISES" if y_val[i] == 1 else "DOES NOT RISE"
    print(f"  Prob: {y_val_proba[i]:.4f} | Actual: {real}")
````

## 3.4 Evaluation Metrics — Is the Model Good?

The model gives a **probability** (0 to 1). We need a **threshold** to decide:
- If probability ≥ threshold → predict "RISES" (1)
- If probability < threshold → predict "DOES NOT RISE" (0)

**Which metrics matter?**

| Metric | Meaning | In trading |
|---------|-------------|------------|
| **Precision** | Of the ones I predicted "RISES", how many actually rose? | If low, you buy junk |
| **Recall** | Of all that actually rose, how many did I detect? | If low, you miss opportunities |
| **F1** | Balance between precision and recall | General metric |
| **AUC** | The model's ability to separate the classes | 0.5 = random, 1.0 = perfect |

In DeepValueAI we prioritize **precision** over recall — we'd rather miss opportunities
than buy stocks that don't rise. **Quality over quantity.**

````python
# Evaluate at different thresholds to understand the tradeoff
print("Threshold | Precision | Recall | F1     | BUY predictions")
print("-" * 65)

for threshold in [0.30, 0.40, 0.50, 0.60, 0.70, 0.80]:
    y_pred = (y_val_proba >= threshold).astype(int)
    n_buy = y_pred.sum()
    if n_buy == 0:
        print(f"  {threshold:.2f}    |    N/A    |  N/A   |  N/A   | {n_buy:,}")
        continue
    prec = precision_score(y_val, y_pred, zero_division=0)
    rec = recall_score(y_val, y_pred, zero_division=0)
    f1 = f1_score(y_val, y_pred, zero_division=0)
    print(f"  {threshold:.2f}    |  {prec:.4f}  | {rec:.4f} | {f1:.4f} | {n_buy:,}")

print("\n→ Notice: a higher threshold = more precision but fewer signals")
print("→ DeepValueAI optimizes the threshold for maximum precision with recall ≥ 25%")
````

## 4.2 Technical Indicators

**Technical indicators** are mathematical formulas that transform price/volume data
into signals that help predict future movements.

DeepValueAI uses 11 technical indicators. Here are the most important ones:

| Indicator | What it measures | Buy signal |
|-----------|----------|-----------------|
| **SMA 200** | Long-term trend | Price near or below the SMA |
| **RSI** | Overbought or oversold? | RSI < 30 (oversold) |
| **Williams %R** | Momentum | Crosses up from -80 |
| **MACD** | Trend change | Histogram crosses from negative to positive |

````python
# Compute technical indicators manually (to understand the logic)

# SMA 200 — average of the last 200 days of closing price
data["SMA_200"] = data["Close"].rolling(window=200).mean()

# RSI — Relative Strength Index (14 days)
delta = data["Close"].diff()
gain = delta.where(delta > 0, 0).rolling(14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(14).mean()
rs = gain / loss
data["RSI"] = 100 - (100 / (1 + rs))

# Distance of the price to the SMA 200 (key DeepValueAI feature)
data["SMA_Distance"] = (data["Close"] - data["SMA_200"]) / data["SMA_200"]

# Visualize
fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# Price + SMA
axes[0].plot(data.index, data["Close"], label="Price", linewidth=1)
axes[0].plot(data.index, data["SMA_200"], label="SMA 200", linewidth=1.5, color="orange")
axes[0].set_title("AAPL — Price vs SMA 200")
axes[0].legend()

# RSI
axes[1].plot(data.index, data["RSI"], color="purple", linewidth=1)
axes[1].axhline(y=70, color="red", linestyle="--", alpha=0.5, label="Overbought (70)")
axes[1].axhline(y=30, color="green", linestyle="--", alpha=0.5, label="Oversold (30)")
axes[1].set_title("RSI (14 days)")
axes[1].legend()
axes[1].set_ylim(0, 100)

# SMA Distance
colors = np.where(data["SMA_Distance"] > 0, "green", "red")
axes[2].bar(
    data.index, data["SMA_Distance"],
    color=colors, width=1,
)
axes[2].axhline(y=0, color="black", linewidth=0.5)
axes[2].set_title("Distance to SMA 200 (% of price)")

plt.tight_layout()
plt.show()
````

## 5.3 Look-Ahead Bias — The #1 Mistake in Finance with ML

**Look-ahead bias** = using information from the future to make decisions in the past.

Example: If on January 1st you use financial data that the company published on March 15th,
you're "cheating" — that data didn't exist on January 1st.

**How does DeepValueAI avoid it?**

1. **Fundamentals**: Uses `publish_date` (actual publication date), not `report_date`
2. **Macro (FRED)**: Uses `realtime_start` (the date the Fed published the data)
3. **Sentiment**: Applies a 1-day lag — Tuesday's news only appears as a feature on Wednesday
4. **Temporal split**: Train always BEFORE validation, validation BEFORE test
5. **Gaps between splits**: A 40-day gap between train and val to avoid label leakage

## 5.5 Backtesting — Would It Have Worked in Real Life?

**Backtesting** simulates how your strategy would have performed if you had used it
in the past. It's the most important test before investing real money.

**DeepValueAI's backtest rules:**
- **Entry**: Model says BUY + price is near the SMA 200
- **Stop Loss**: If it loses 12% from the purchase → sell everything
- **Trailing Stop**: If it rises 5% and then falls 8% from the peak → sell
- **Partial Take Profit**: If it rises 15% → sell 33% of the position
- **Maximum time**: If 120 days pass without triggering anything → sell
- **Diversification**: Maximum 10 open positions, maximum 15% per ticker

````python
# Run a mini-backtest with 10 tickers (the full one takes ~20 min)
from core.backtesting_engine import run_backtest

test_tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA",
                "JPM", "JNJ", "V", "PG", "META"]

print(f"Running backtest with {len(test_tickers)} tickers...")
print("(This downloads data and simulates trading day by day — it may take 2-3 minutes)\n")

result = run_backtest(
    tickers=test_tickers,
    start_date="2022-01-01",
    end_date="2025-12-31",
    initial_capital=100_000.0,
)

# Show results
m = result.metrics
print("=" * 50)
print("BACKTEST RESULTS")
print("=" * 50)
print(f"  Total return:       {m['total_return']:.2%}")
print(f"  Annualized return:  {m['annualized_return']:.2%}")
print(f"  Sharpe Ratio:       {m['sharpe_ratio']:.2f}")
print(f"  Max Drawdown:       {m['max_drawdown']:.2%}")
print(f"  Win Rate:           {m['win_rate']:.2%}")
print(f"  Profit Factor:      {m['profit_factor']:.2f}")
print(f"  Total Trades:       {m['num_trades']}")
print(f"  Equity: $100,000 → ${result.equity_curve.iloc[-1]:,.0f}")
````

````python
# Drift Monitoring — detect when the model degrades
import numpy as np

from core.drift_monitor import check_prediction_stability

# Simulate predictions from two different periods
np.random.seed(42)
predictions_old = np.random.beta(2, 5, size=200)    # original distribution
predictions_new = np.random.beta(2, 5, size=200)    # similar distribution

# Compare — is there drift?
result = check_prediction_stability(predictions_old, predictions_new)
print("=== Stability Test (similar distributions) ===")
print(f"  KS Statistic: {result['ks_statistic']:.4f}")
print(f"  p-value:      {result['p_value']:.4f}")
print(f"  Drift:        {'⚠️ YES' if result['significant_drift'] else '✅ NO'}")

# Now simulate a real drift
predictions_drifted = np.random.beta(5, 2, size=200)  # VERY different distribution

result2 = check_prediction_stability(predictions_old, predictions_drifted)
print("\n=== Stability Test (VERY different distributions) ===")
print(f"  KS Statistic: {result2['ks_statistic']:.4f}")
print(f"  p-value:      {result2['p_value']:.4f}")
print(f"  Drift:        {'⚠️ YES' if result2['significant_drift'] else '✅ NO'}")
print("\n→ When there's drift, it's time to retrain the model.")
````
